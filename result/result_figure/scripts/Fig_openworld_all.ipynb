```python
import numpy as np
def create_multi_bars(labels, datas, tick_step=1,legend=[], group_gap=0.2, bar_gap=0,filename=''):
    '''
    labels : x
    datas labels
    tick_step x1tick_stepx
    group_gap : 
    bar_gap 0
    '''
    plt.figure(figsize=(6, 6))
    # fig, ax = plt.subplots()
    # ticksx
    # plt.ylim(4, 4.2)

    patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
    ticks = np.arange(len(labels)) * tick_step
    # group_num
    group_num = len(datas)
    # group_widthgroup_gap 
    group_width = tick_step - group_gap
    # bar_spanx
    bar_span = group_width / group_num
    # bar_width
    bar_width = bar_span - bar_gap
    # baseline_xxbar_span
    baseline_x = ticks - (group_width - bar_span) / 2
    for index, y in enumerate(datas):
        plt.bar(baseline_x + index*bar_span, y, bar_width,hatch=patterns[index])
    plt.ylabel('NT2T rate',fontdict=font)

    # plt.title('multi datasets')
    # xx
    plt.xticks(ticks, labels,size = 24)
    # plt.xlabel('Open-Domain Chatbot Models',fontdict=font)  # x
    plt.yticks(size = 24)
    plt.legend(['Ehanced with Org','Default'],fontsize=20,prop={'size':20},loc='lower left')
    # plt.savefig(filename, bbox_inches='tight',  pad_inches = 0)
    # plt.hlines(4.05, -0.5, 4.5, linewidth = 4,color = 'grey',linestyles ='--')
    plt.savefig(os.path.join(figure_path,filename), dpi=300,bbox_inches='tight',  pad_inches = 0)
    
    plt.show()
new_combined_data=[
    numpy.add(all_data_list_e[2],all_data_list_e[4])*0.5,
    numpy.add(all_data_list_e[0],all_data_list_e[3])*0.5
]
create_multi_bars(['DialoGPT-L', 'BB-s', 'BB-L'],new_combined_data,legend=[],bar_gap=0.02,filename='Fig4_b.pdf')
```

```python
import numpy as np
def create_multi_bars(labels, datas, tick_step=1,legend=[], group_gap=0.2, bar_gap=0,filename=''):
    '''
    labels : x
    datas labels
    tick_step x1tick_stepx
    group_gap : 
    bar_gap 0
    '''
    plt.figure(figsize=(6, 6))
    # fig, ax = plt.subplots()
    # ticksx
    # plt.ylim(4, 4.2)

    patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
    ticks = np.arange(len(labels)) * tick_step
    # group_num
    group_num = len(datas)
    # group_widthgroup_gap 
    group_width = tick_step - group_gap
    # bar_spanx
    bar_span = group_width / group_num
    # bar_width
    bar_width = bar_span - bar_gap
    # baseline_xxbar_span
    baseline_x = ticks - (group_width - bar_span) / 2
    for index, y in enumerate(datas):
        plt.bar(baseline_x + index*bar_span, y, bar_width,hatch=patterns[index])
    plt.ylabel('NT2T rate',fontdict=font)

    # plt.title('multi datasets')
    # xx
    plt.xticks(ticks, labels,size = 24)
    # plt.xlabel('Open-Domain Chatbot Models',fontdict=font)  # x
    plt.yticks(size = 24)
    plt.legend(['Ehanced with Prompt','Default'],fontsize=20,prop={'size':20},loc='lower left')
    # plt.savefig(filename, bbox_inches='tight',  pad_inches = 0)
    # plt.hlines(4.05, -0.5, 4.5, linewidth = 4,color = 'grey',linestyles ='--')
    plt.savefig(os.path.join(figure_path,filename), dpi=300,bbox_inches='tight',  pad_inches = 0)
    
    plt.show()
new_combined_data=[
    all_data_list_e[0],
    numpy.add(all_data_list_e[1],all_data_list_e[2])*0.5
]
create_multi_bars(['DialoGPT-L', 'BB-s', 'BB-L'],new_combined_data,legend=[],bar_gap=0.02,filename='Fig4_d.pdf')
```
